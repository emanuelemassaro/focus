```python
## Cantonal Boundaries
pathShape = prefix+'Dropbox (MIT)/HERUS_Lab/SwissDateset/cb/'
sf = shapefile.Reader(pathShape+"cantonalBoundaries.shp")
recs    = sf.records()
shapes  = sf.shapes()
Nshp    = len(shapes)
```

```python
fig,ax = plt.subplots(figsize=(18,11))
cout = 0
tot = 0
l = []
for nshp in xrange(Nshp):
    if recs[cout][18] not in l:
        l.append(recs[cout][18])
        tot+=1
    cout+=1
    pts     = array(shapes[nshp].points)
    prt     = shapes[nshp].parts
    par     = list(prt) + [pts.shape[0]]
    bbPath = mplPath.Path(pts)
    cc = 'gray'
    ptchs   = []
    for pij in xrange(len(prt)):
        
        ptchs.append(Polygon(pts[par[pij]:par[pij+1]]))
        ax.add_collection(PatchCollection(ptchs,facecolor=cc,edgecolor='k', linewidths = 1))
ax.set_xlim(5.5, 11)
ax.set_ylim(45.5,48)
ax.set_xlabel('Latitude',size = 30)
ax.set_ylabel('Longitude',size= 30)
ax.set_title('Cantonal Boundaries',size= 30)
```

```python
xdata = tst['women'] + tst['men']
ydata = tst['Road length (km)']
init_vals = [0.01, 0.8]
best_vals, covar = curve_fit(powerlaw, xdata, ydata, p0=init_vals)


ypred = powerlaw(xdata, best_vals[0], best_vals[1])
fig,ax = plt.subplots(figsize=(6,6))
plt.plot(xdata, ydata, lw = 0, marker= "o")
plt.plot(xdata, ypred)
ax.set_xscale("log", nonposx='clip')
ax.set_yscale("log", nonposy='clip')
ax.set_xlabel('Inhabitants',size = 18)
ax.set_ylabel('Road Length, '+r'$km$',size = 18)
string1 = r'$\beta$ = %.2f'%best_vals[1]
string2 = r'$R^2$ = %.2f'%(r2_score(ydata, ypred))

ax.annotate(string1,
            xy=(700, 300), xycoords='figure pixels')
ax.annotate(string2,
            xy=(700, 200), xycoords='figure pixels')
```
