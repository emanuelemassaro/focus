```python
def returnCantonData(zips, name, mobiliereData):
    tmp = zips[zips.iloc[:,4] == name].reset_index(drop=True)
    tmp = pd.DataFrame(tmp.iloc[:,1])
    tmp.columns = ['ZIP']
    df = pd.merge(mobiliereData, tmp)  ## Data Canton Geneva
    x = df.filter(['ZIP', var])
    return x
def returnCityData(zips, bfs, mobiliereData):
    tmp = zips[zips.iloc[:,8]==bfs].reset_index(drop=True)
    tmp = pd.DataFrame(tmp.iloc[:,1])
    tmp.columns = ['ZIP']
    df = pd.merge(mobiliereData, tmp)  ## Data Canton Geneva
    x = df.filter(['ZIP', var])
    return x
def plot(x, vmin, vmax, title, column):
    bg_color = 'white'
    fg_color = 'white'
    fig, ax = plt.subplots(facecolor=bg_color, figsize=(5,5))
    x.plot(ax = ax, column= column ,cmap='Blues', scheme='quantiles', edgecolor='darkgrey')
    ax.axis('off');
    #ax.set_title(title, fontdict={'fontsize': '18', 'fontweight' : '3', 'color': 'black'});
    sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
    sm._A = []
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="2%", pad=0.05)
    cbar = fig.colorbar(sm, cax=cax)
    return fig
```
